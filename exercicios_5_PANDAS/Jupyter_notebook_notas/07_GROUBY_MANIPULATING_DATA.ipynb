```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks['beer_servings'].mean()
```

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean()
```

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].max()
```

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].min()
```

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].agg(['count', 'max', 'min', 'mean'])
```

```python
drinks.groupby('continent').agg(['count', 'max', 'min', 'mean'])
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].agg(['count', 'max', 'min', 'mean']).plot(kind='bar')
```
